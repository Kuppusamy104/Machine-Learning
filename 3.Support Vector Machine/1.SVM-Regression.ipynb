```python
import pandas as pd
```

```python
dataset=pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset[['State_Florida','State_New York']] = dataset[['State_Florida','State_New York']].astype(int) # Change the column dtype to int after applying get_dummies
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']]
```

```python
dependent=dataset[["Profit"]]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
x_train
```

```python
# data pre processing -standardization 
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train) #fit_transform => (x-mean)/std deviation for each data point
x_test=sc.transform(x_test) #transform -> applying training data standardization value for test data as well while doing cal's
```

```python
x_train
```

```python
#ref : https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVR.html
#default kernel value is "rbf"
from sklearn.svm import SVR
#regressor=SVR(kernel="rbf",C=1.0)  
#regressor=SVR(kernel="linear",C=1.0)  
#regressor=SVR(kernel="poly",C=1.0) 
#regressor=SVR(kernel="sigmoid",C=10) 
regressor=SVR(kernel="rbf",C=1000)
regressor.fit(x_train,y_train)

#C-Regularization parameter. default is 1.0
#regressor=SVR(kernel="linear")  
#kernel{‘linear’, ‘poly’, ‘rbf’, ‘sigmoid’, ‘precomputed’} 
# creatingb the SVM model using linear format (kernel="linear")
#kernel="rbf" ref representing non linear (radial basis function)
```

```python
bais=regressor.intercept_ 
bais #origin value 
```

```python
regressor.n_support_ # data created using how many support vector
#this model created using 34 dat points
```

```python
regressor.support_
```

```python
y_pred = regressor.predict(x_test) 
```

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test,y_pred)
```
