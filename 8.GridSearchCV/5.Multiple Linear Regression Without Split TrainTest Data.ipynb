```python
import pandas as pd
dataset = pd.read_csv("insurance_pre.csv")
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset[['sex_male','smoker_yes']] = dataset[['sex_male','smoker_yes']].astype(int) # Change the column dtype to int after applying get_dummies
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
dependent=dataset[['charges']]
```

```python
#from sklearn.model_selection import train_test_split
#x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression
param_grid= {'fit_intercept': [True,False],
             'copy_X': [True,False],
             'n_jobs': [None,5,10]}

grid= GridSearchCV(LinearRegression(), param_grid, refit=True, verbose=3, n_jobs=-1) #estimator=LinearRegression()
grid.fit(independent,dependent) # model creation using grid serch for linear using above setting
```

```python
#print best parameter after tuning
re=grid.cv_results_
print("The R_score value for best parameter {}:".format(grid.best_params_))
```

```python
y_pred = grid.predict(independent) 
```

```python
from sklearn.metrics import r2_score
r2_score = r2_score(dependent,y_pred)
r2_score
```
